# Analyzing CIA Factbook Data

```python
import pandas as pd
import sqlite3
```

```python
q1 = "SELECT * FROM sqlite_master WHERE type='table';"
conn = sqlite3.connect("factbook.db")
cursor = conn.cursor()
cursor.execute(q1).fetchall()
```

```python
pd.read_sql_query(q1, conn)
```

```python
peek_data = "select * from facts limit 10"
pd.read_sql_query(peek_data, conn)
```

## Querying Summary Population Statistics

```python
summary_stats = '''
select min(population) Minimum_Population, 
max(population) Maximum_Population, 
min(population_growth) Minimum_Population_Growth, 
max(population_growth) Maximum_Population_Growth from facts
'''
pd.read_sql_query(summary_stats, conn)
```

## Identifying Outliers in the Data

```python
min_pop = 'select * from facts where population == (select min(population) from facts)'
pd.read_sql_query(min_pop, conn)
```

```python
max_pop = 'select * from facts where population == (select max(population) from facts)'
pd.read_sql_query(max_pop, conn)
```

We can identify that according to CIA records, Antarctica has the lowest population with stagnant population growth. This data is consistent with CIA records of Antarctica (https://www.cia.gov/library/publications/the-world-factbook/geos/ay.html). However, our maximum population value is actually the CIA's estimated global population which is a little over 7.2 billion people. 

## Visualizing Populations

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

hist_data = '''
select population, population_growth, birth_rate, death_rate
from facts
'''
pd.read_sql_query(hist_data, conn).hist(ax=ax)
```

## Countries with Highest Population Density

```python
q7 = "select name, cast(population as float)/cast(area as float) density from facts order by density desc limit 20"
pd.read_sql_query(q7, conn)
```

```python
q7 = '''select population, population_growth, birth_rate, death_rate
from facts
where population != (select max(population) from facts)
and population != (select min(population) from facts);
'''
pd.read_sql_query(q7, conn)
```
